```python
import os
import glob
from xml.etree import ElementTree as ET

# Path to the folder containing your SVG files
folder_path = "./"
stroke_wdith = "1.5"

# Iterate through all .svg files in the folder
for svg_file in glob.glob(os.path.join(folder_path, "*.svg")):
    try:
        tree = ET.parse(svg_file)
        root = tree.getroot()

        # Update stroke-width in root if present
        if 'stroke-width' in root.attrib:
            root.attrib['stroke-width'] = stroke_wdith

        # Also update stroke-width in nested elements (optional but thorough)
        for elem in root.iter():
            if 'stroke-width' in elem.attrib:
                elem.attrib['stroke-width'] = stroke_wdith

        # Save changes back to file
        tree.write(svg_file)
        print(f"Updated: {svg_file}")
    except ET.ParseError as e:
        print(f"Failed to parse {svg_file}: {e}")
```
